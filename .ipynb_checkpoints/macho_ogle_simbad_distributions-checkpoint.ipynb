```python
font = {'family' : 'normal',
        'size'   : 12,
        'weight' : 'bold'}

matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10)
matplotlib.rcParams["ps.useafm"] = True
matplotlib.rcParams["pdf.use14corefonts"] = True
matplotlib.rcParams["text.usetex"] = True
```

```python
def three_cell_plot(x1, x2, x3, y1, y2, y3, xlabel=None, xlims=None, ylabel=None, ylims=None, txt1=None, txt2=None, txt3=None, inverted=False, outfile=None, vertline=None, fsize=16):
    colors = ["b","r","g","y","purple","orange","black"]
    
    fig = plt.figure(figsize=(6,6))
    fig.subplots_adjust(left=0.1, right=0.95, top=0.95, bottom=0.1, wspace=0, hspace=0)

    ax = plt.subplot(221)
    for ii in range(len(x1)):
        ax.scatter(x1[ii], y1[ii], c=colors[ii], s=1, edgecolor="None")
    if vertline:
        ax.plot([vertline, vertline], [ylims[0], ylims[1]], color='k', linestyle="--")
    if txt1:
        ax.text(0.9, 0.9, txt1, transform=ax.transAxes, horizontalalignment="right", fontsize=14)
    ax.set_xlim(xlims[0], xlims[1])
    ax.set_ylim(ylims[0], ylims[1])
    ax.set_ylabel(ylabel, fontsize=fsize)
    if inverted:
        ax.invert_yaxis()
    ax.get_xaxis().set_ticklabels([])
    ax.minorticks_on()

    ax = plt.subplot(223)
    for ii in range(len(x2)):
        ax.scatter(x2[ii], y2[ii], c=colors[ii], s=1, edgecolor="None")
    if vertline:
        ax.plot([vertline, vertline], [ylims[0], ylims[1]], color='k', linestyle="--")
    if txt2:
        ax.text(0.9, 0.9, txt2, transform=ax.transAxes, horizontalalignment="right", fontsize=14)
    ax.set_xlim(xlims[0], xlims[1])
    ax.set_ylim(ylims[0], ylims[1])
    ax.set_xlabel(xlabel, fontsize=fsize)
    ax.set_ylabel(ylabel, fontsize=fsize)
    if inverted:
        ax.invert_yaxis()
    ax.minorticks_on()

    ax = plt.subplot(224)
    for ii in range(len(x3)):
        ax.scatter(x3[ii], y3[ii], c=colors[ii], s=1, edgecolor="None")
    if vertline:
        ax.plot([vertline, vertline], [ylims[0], ylims[1]], color='k', linestyle="--")
    if txt3:
        ax.text(0.9, 0.9, txt3, transform=ax.transAxes, horizontalalignment="right", fontsize=14)
    ax.set_xlim(xlims[0], xlims[1])
    ax.set_ylim(ylims[0], ylims[1])
    ax.set_xlabel(xlabel, fontsize=fsize)
    ax.get_yaxis().set_ticklabels([])
    if inverted:
        ax.invert_yaxis()
    ax.minorticks_on()
    if outfile:
        plt.savefig(outfile)
    
    plt.show()
```

```python
def cumsum(arr):
    the_sum = np.zeros(len(arr))
    the_sum[0] = arr[0]
    for ii in range(1, len(arr)):
        the_sum[ii] = arr[ii] + the_sum[ii-1]
        
    return np.array(the_sum)
```
